# Monte-Carlo Algorithm for Estimating $\pi$ in Python
This notebook will provide a demonstration of how to estimate $\pi$ using a Monte-Carlo algorithm. 

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

First, we choose the number of tosses: 

```python
n_tosses = 10000
```

Then, we randomly generate x and y corrdinates between 0 and 1: 

```python
x = np.random.rand(n_tosses)
y = np.random.rand(n_tosses)
```

Now, we count the number of points inside the circle: 

```python
n_in_circle = 0
for i in range(0, n_tosses - 1):
    if (x[i]**2 + y[i]**2 <= 1):
        n_in_circle += 1
```

And here is our estimated $\pi$: 

```python
pi_estimate = 4 * (n_in_circle / n_tosses)
print("π =", pi_estimate)
```

The percent error is:

```python
print(100 * math.fabs(math.pi - pi_estimate) / math.pi, "%")
```

```python
circle_x = x[np.sqrt(x**2 + y**2) <= 1]
circle_y = y[np.sqrt(x**2 + y**2) <= 1]
```

```python
fig = plt.figure()
plot = fig.add_subplot(111)
plot.scatter(x, y, marker='.', color='blue')
plot.scatter(circle_x, circle_y, marker='.', color='red')

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)

plot.plot(x, y, color='black')

plot.set_aspect(1.0)

plt.rcParams['figure.figsize'] = [20, 20]
plt.show()
```
